```python
# Getting null columns and count of them
null_counts = df.isnull().sum()
if null_counts.any():
    print(null_counts[null_counts > 0])
```

```python
# Dropping columns which contains so many missing values
df1 = df.drop(["Alley","MasVnrType","PoolQC","Fence","MiscFeature"],axis=1)
```

```python
# Iterate over numeric columns only
for col in df1.select_dtypes(include=[np.number]).columns:
    if col != 'SalePrice':  # Exclude 'SalePrice' itself from comparison
        corr_value = df1['SalePrice'].corr(df1[col])
        print(f"The {col} corr is : {corr_value}")
```

```python
# Dropping columns with low correlation
lower_bound = -0.5
upper_bound = 0.5

# Find columns whose correlation is between -0.5 and 0.5
cols_to_drop = []

for col in df1.select_dtypes(include=[np.number]).columns:
    if col != 'SalePrice':
        corr_value = df1['SalePrice'].corr(df1[col])
        if lower_bound < corr_value < upper_bound:
            cols_to_drop.append(col)
            print(f"Dropping {col} with correlation: {corr_value}")

# Create a new DataFrame by dropping the selected columns
df2 = df1.drop(cols_to_drop, axis=1)
```

```python
ohe = OneHotEncoder()
for col in df2.columns:
    if df2[col].dtype == 'object':  # Check if the column is categorical
        if len(df2[col].unique()) <= 3:  # Apply ordinal encoding for columns with 3 or fewer unique values
            df2[[col]] = oe.fit_transform(df2[[col]])
        else:  # Apply one-hot encoding for columns with more than 3 unique values
            ohe_result = ohe.fit_transform(df2[[col]]).toarray()  # Convert sparse matrix to dense
            # Create a DataFrame for the one-hot encoded result
            ohe_df = pd.DataFrame(ohe_result, columns=[f"{col}_{cat}" for cat in ohe.categories_[0]])
            # Concatenate the new one-hot encoded columns with the original DataFrame
            df2 = pd.concat([df2.reset_index(drop=True), ohe_df.reset_index(drop=True)], axis=1)
            # Drop the original column after one-hot encoding
            df2.drop(columns=[col], inplace=True)
df2.shape
```
